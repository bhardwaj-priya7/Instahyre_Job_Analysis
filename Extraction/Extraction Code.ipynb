```python
import numpy as np
import pandas as pd
from time import sleep
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

### For Extracting The Job Pages Links

```python
# html = driver.find_element(By.TAG_NAME, 'html')
# html.send_keys(Keys.END)
page=[]
nextpage='//*[@id="job-function-page"]/div[2]/div/div[1]/div[1]/div[11]/li[12]'
driver=webdriver.Chrome(options=webdriver.ChromeOptions())
driver.maximize_window()
driver.get('https://www.instahyre.com/search-jobs/')
sleep(2)
for PgNum in range(1,101):
    a=driver.find_elements(By.CLASS_NAME,"text-link")
    for elmnt in a:
        link=elmnt.get_attribute('href')
        if link not in page:
            page.append(link)
            sleep(1)
    driver.find_element(By.XPATH,nextpage).click()
    print(f'page number {PgNum} is done')
    sleep(5)
driver.quit()
```

```python
len(page)
```

```python
pd.DataFrame(page).to_csv("job webLinks.csv")
```

### For Extracting All Elements of Job Page

```python
filepath=r"./Webpages.csv"
```

```python
df=pd.read_csv(filepath)
df
```

```python
final=[]
driver=webdriver.Chrome(options=webdriver.ChromeOptions())
driver.maximize_window()
driver.implicitly_wait(5)
pgcnt=1
for pg in df['0']:
    driver.switch_to.window(driver.window_handles[-1])
    driver.get(pg)
    try :
        LIn=driver.find_element(By.ID, 'employer-linkedin').get_attribute('href')
    except :
        LIn='-----'
    try :
        sleep(2)
        intern=driver.find_element(By.XPATH,'//*[@id="employer-profile"]/div[2]/div/div/div[2]/div[2]/div[1]/div/h2')
        if "Internship Details" in intern.text:
            inv='Internship'
            exper=0    
    except:
        sleep(2)
        inv="Full Time"
        exp= driver.find_element(By.XPATH,'//*[@id="floating-header"]/div[1]/div/span[2]')
        exper= exp.text
    Company= driver.find_element(By.XPATH,'//*[@id="employer-profile"]/div[2]/div/div/div[1]/div[2]/h5')
    estab_year=driver.find_element(By.XPATH,'//*[@id="employer-profile"]/div[2]/div/div/div[1]/div[2]/div[1]')
    emp_cnt=driver.find_element(By.XPATH,'//*[@id="employer-profile"]/div[2]/div/div/div[1]/div[2]/div[2]')
    location =driver.find_element(By.XPATH,'//*[@id="floating-header"]/div[1]/div/span[1]')
    designation =driver.find_element(By.CLASS_NAME,"profile-info")
    skills = driver.find_element(By.XPATH,'//*[@id="job-description"]/div/div/div[1]')
    hr_name= driver.find_element(By.CLASS_NAME,"rec-name")
    driver.implicitly_wait(5)
     
    final.append({'Company Name':Company.text.replace(' at a glance',''),
                  "Estab_year":estab_year.text.replace('Founded in ',''),
                  "Employees_count":emp_cnt.text.replace(' employees',''),
                  "Location":location.text,
                  "Designation":designation.text.split('\n')[0],
                  "Involvement": inv,
                  "Skills":','.join(skills.text.split('\n')),
                  "HR_Name":hr_name.text,
                  "Experience":exper,
                  "linkedIN" :LIn,
                  })
    driver.implicitly_wait(5)
    # opening a new tab in chrome
    driver.execute_script("window.open('');")
    driver.implicitly_wait(5)
    #switching to last tab and 
    driver.switch_to.window(driver.window_handles[0])
    driver.implicitly_wait(5)
    # closing the tab 
    driver.close()
    print(f'Page {pgcnt} Done')
    pgcnt+=1
driver.quit()
```

```python
Final_df=pd.DataFrame(final)
Final_df
```
